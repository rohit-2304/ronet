```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/housing/Housing.csv')
df.head()
```

```python
df['mainroad'] = df['mainroad'].map({'yes': 1, 'no': 0})
df['guestroom'] = df['guestroom'].map({'yes': 1, 'no': 0})
df['basement'] = df['basement'].map({'yes': 1, 'no': 0})
df['hotwaterheating'] = df['hotwaterheating'].map({'yes': 1, 'no': 0})
df['airconditioning'] = df['airconditioning'].map({'yes': 1, 'no': 0})
df['prefarea'] = df['prefarea'].map({'yes': 1, 'no': 0})
```

```python
df = pd.get_dummies(df, columns=['furnishingstatus'], drop_first=True)

df['furnishingstatus_semi-furnished'] = df['furnishingstatus_semi-furnished'].map({False:1, True:0})
df['furnishingstatus_unfurnished'] = df['furnishingstatus_unfurnished'].map({False:1, True:0})
```

```python
df = df.dropna()

print("\nAfter preprocessing:")
df.head()
```

```python
df.to_csv('data/housing/cleaned.csv')
```
